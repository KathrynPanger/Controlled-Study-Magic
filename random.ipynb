<h1>Test for Spell Efficacy</h1>
<h2>Using magic to create statistical bias in a randomly generated list of inegers between 0 and 100</h2>
This code randomly generates two list. The first list is generated without a magic spell cast upon the computer/notebook to create bias in the numbers. The experimental group is generated after th magic spell for bias has been cast. A regression test is conducted on the test and control groups to see if there is a statsitically significant difference between the values of the two lists at 95% confidence. The results represent the chances of getting the result that we did by random chance [h0] and this is compared to the chances of getting it through the causal mechanism of spell-casting [h1]. This study is reproduceable. Simply run the cells again, casting the magic spell at the correct time. 

```python
#Generate Control Group
control=[]
for i in range(0,60):
    d = random.randint(0,100)
    control.append(d)
```

<h2>Cast Spell Now</h2>

```python
#Generate Test Group
test = []
for i in range(0,60):
    d = random.randint(0,100)
    test.append(d)
```

```python
study={"Control": control, "Test":test}
```

```python
df=pd.DataFrame(study)
```

```python
reg(df,"Control", "Test")
```

<h2>Interpreting Results</h2>
Now, examine the P-score (P>|t|) for the variable "Test" (not for intercept, this does not tell us anything). This number represnets how confident we can be that casting a spell creates bias in our number generation. If P<0.05, we can be 95% confident that the two groups are distinct, and that our results are not due to random chance. The residuals plot visualizes the results.
